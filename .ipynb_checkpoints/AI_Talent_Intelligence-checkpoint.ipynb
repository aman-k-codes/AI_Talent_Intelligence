```python
import fitz
import spacy

nlp = spacy.load("en_core_web_sm")

def extract_text_from_pdf(path):
    pages = fitz.open(path)
    text = ''
    links = []

    for page in pages:
        text += page.get_text()
        for url in page.get_links():
            if 'uri' in url:
                links.append(url['uri'])

    return text, links
```

```python
rawText, rawLinks = extract_text_from_pdf("Aman_Kumar_Resume_entry_lvl_mld.pdf")
```

```python
def extract_candidate_name(text):
    doc = nlp(text)

    for ent in doc.ents:
        if ent.label_ == "PERSON":
            return ent.text

    return None
```
